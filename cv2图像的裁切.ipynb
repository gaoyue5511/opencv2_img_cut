```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
img = cv2.imread('timg.jpg') # 导入图片
plt.imshow(img,'gray')
plt.show()                   # 展示图片

batch_num_x = 10 #竖着分多少份
batch_num_y = 16 #横着分多少份


img_shape = np.shape(img)  # 以下是对原图片进行分割计算 本程序的功能是按照用户想把一张图片分为多少份，依据用户输入的x，y来分割。

x = int(img_shape[0])
y = int(img_shape[1])

batch_x = int(x/batch_num_x)
batch_y = int(y/batch_num_y)


# 以下是进行突变的裁切，然后把裁切下的图片，追加到imgs列表里

imgs = []

for idx in range(batch_x,x+1,batch_x):
    for idy in range(batch_y,y+1,batch_y):
        imgs.append(img[idx-batch_x:idx,idy-batch_y:idy])

        
        
        
#图片全部存在imgs里面，可以直接添加下标，输出看下。

#plt.imshow(imgs[50],'gray')
#plt.show()


# 下面是创建文件夹的准备操作
cur_path = os.path.abspath(os.curdir) # 定位当前文件夹

# 以下是依照我的文件路径，分别把样本图片放到各自的文件夹中。

img2file = []
for file_path_id_x in range(batch_num_x):
    file_path_name_a = str(file_path_id_x)
    file_path_a = cur_path+"\\data3\\"+file_path_name_a+"\\"
    os.makedirs(file_path_a)
    for file_path_id_y in range(batch_num_y):
        file_path_name_b = str(file_path_id_y)
        file_path =file_path_a + file_path_name_b +"M"+".jpg"
        page = file_path_id_y+file_path_id_x*batch_num_y
        # print (page,',',end='')    测试用的
        img2file = imgs[page]
        cv2.imencode('.jpg', img2file)[1].tofile(file_path)  #这个是输出文件的语句，已经注释掉了，用时去掉注释。


#cv2.imwrite('frist.jpg',imgs)  # 之前使用这个来输出图片的，后来从网上查，不好用。
#cv2.imencode('.jpg', imgs)[1].tofile(file_path)  # 然后就用了这个方法来输出图片



```
