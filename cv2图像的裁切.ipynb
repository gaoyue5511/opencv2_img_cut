```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
img = cv2.imread('timg.jpg') # 
plt.imshow(img,'gray')
plt.show()                   # 

batch_num_x = 10 #
batch_num_y = 16 #


img_shape = np.shape(img)  #  xy

x = int(img_shape[0])
y = int(img_shape[1])

batch_x = int(x/batch_num_x)
batch_y = int(y/batch_num_y)


# imgs

imgs = []

for idx in range(batch_x,x+1,batch_x):
    for idy in range(batch_y,y+1,batch_y):
        imgs.append(img[idx-batch_x:idx,idy-batch_y:idy])

        
        
        
#imgs

#plt.imshow(imgs[50],'gray')
#plt.show()


# 
cur_path = os.path.abspath(os.curdir) # 

# 

img2file = []
for file_path_id_x in range(batch_num_x):
    file_path_name_a = str(file_path_id_x)
    file_path_a = cur_path+"\\data3\\"+file_path_name_a+"\\"
    os.makedirs(file_path_a)
    for file_path_id_y in range(batch_num_y):
        file_path_name_b = str(file_path_id_y)
        file_path =file_path_a + file_path_name_b +"M"+".jpg"
        page = file_path_id_y+file_path_id_x*batch_num_y
        # print (page,',',end='')    
        img2file = imgs[page]
        cv2.imencode('.jpg', img2file)[1].tofile(file_path)  #


#cv2.imwrite('frist.jpg',imgs)  # 
#cv2.imencode('.jpg', imgs)[1].tofile(file_path)  # 



```
